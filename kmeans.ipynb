```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    'age': [10, 20, 30, 40, 50, 60,23,55,24,78],
    'income': [32, 44, 26, 44, 12, 34,44,35,66,43],
    'score': [12, 23, 45, 3, 65, 2,56,23,53,61]
}

df=pd.DataFrame(data)
df
```

```python
x=df.iloc[:,[1,2]].values
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_init=10,max_iter=300,random_state=0,n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
kmeans=KMeans(n_init=10,max_iter=300,random_state=0,n_clusters=3,init='k-means++')
y_hc=kmeans.fit_predict(x)
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],c='red',label='cluster1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],c='blue',label='cluster2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],c='yellow',label='cluster3')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, c = 'yellow', label
= 'Centroids')
plt.legend(fontsize=16)
plt.grid(True)
plt.show()
```
